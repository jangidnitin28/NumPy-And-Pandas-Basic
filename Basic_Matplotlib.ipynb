```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = np.linspace(0,10,10)
data1
data2 = data1**3
data2
```

```python
plt.plot(data1,data2)
plt.title("Nitin Data")
plt.xlabel("Discount")
plt.ylabel("Sell")
```

```python
plt.subplot(2,2,1)
plt.plot(data1,data2)
plt.subplot(2,2,2)
plt.plot(data1,data2)
plt.subplot(2,2,3)
plt.plot(data2,data1)
plt.subplot(2,2,4)
plt.plot(data1,data2)
```

```python
fig = plt.figure(figsize=(5,5),dpi=100)
axes1 = fig.add_axes([.1,.1,1,1])
axes1.plot(data1,data2)

axes2 = fig.add_axes([.2,.2,.4,.4])
axes2.plot(data1,data2)
```

```python
plt.scatter(data1,data2)
```

```python
plt.hist(data2)
```

```python
plt.boxplot(data1)
```

```python
fig = plt.figure()
axis = fig.add_axes([.1,.1,.8,.8])
axis.plot(data1,data2)
 
plt.savefig("basic.png")
```
